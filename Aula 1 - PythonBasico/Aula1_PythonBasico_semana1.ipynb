### Estruturas de controle

```python
'''
numero <-- leia um nmero do teclado
se nmero for par
    se numero > 10
        imprima sim
    seno
        imprima no
do contrrio, se for mpar
    se numero < 50
        imprima sim
    seno
        imprima no
'''
```

```python
numero = int(input('Digite um numero: '))
if numero % 2 == 0 and numero != 0:
  if numero > 10:
    print ('sim')
  else:
    print('nao')
elif numero == 0:
  print('numero  igual a zero')
else:
  if numero < 50:
    print('sim')
  else:
    print('nao')
```

## For e While

Lao de repetio para fazer uma contagem. 

```python
x = 0
for a in range(9,-1,-1):
  x = x + 1
  print(f'quando o contador for {a} x vai ser {x}')  
```

fazer uma contagem ao contrrio

```python
for c in range(6,0,-1):
  print(c)
print('fim')
```

o ultimo argumento define o que vai ser feito

```python
for c in range(0,8,2):
  print(c)
print('fim')
```

```python
n = int(input('Digite um numero: '))
for c in range (0,n):
  print(c)
print('fim')
```

## While

```python
'''for c in range (1,10):
  print(c)
print('fim')'''


c=1 
while c < 10:
  print(c)
  c += 1
print('fim')
```

```python
n = 1
while n != 0:
  n = int(input('Digite um valor: '))
print('Fim')
```

```python
r = 'S'
while r == 'S':
    n = int(input('Digite um valor: '))
    r = str(input('Quer continuar? [S/N] ')).upper()
print('Fim')
```

```python
n = 1
par = 0
impar = 0
while n != 0:
  n = int(input('Digite um valor: '))
  if n != 0:
    if n % 2 == 0 and n != 0:
      par += 1
    else:
      impar += 1
print('Voce digitou {} nmeros pares e {} numeros impares'.format(par,impar))
```
